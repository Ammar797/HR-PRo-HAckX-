```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("/content/CompactAttrition.csv")

# Group data by Attrition
attrition_grouped = data.groupby('Attrition')

# Calculate average values for selected features
selected_features = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome']
avg_values = attrition_grouped[selected_features].mean()

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
fig.suptitle('Average Feature Values by Attrition', fontsize=20)

for i, feature in enumerate(selected_features):
    ax = axes[i//4, i%4]
    avg_values[feature].plot(kind='bar', ax=ax, color=['skyblue', 'salmon'])
    ax.set_title(feature)
    ax.set_xlabel('Attrition')
    ax.set_ylabel('Average Value')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("/content/CompactAttrition.csv")

# Group data by age and attrition
grouped_data = data.groupby(['Age', 'Attrition']).agg({'MonthlyIncome': 'mean'}).reset_index()

# Filter data for visualization
attrition_yes = grouped_data[grouped_data['Attrition'] == 'Yes']
attrition_no = grouped_data[grouped_data['Attrition'] == 'No']

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(attrition_yes['Age'], attrition_yes['MonthlyIncome'], label='Attrition: Yes', marker='o')
plt.plot(attrition_no['Age'], attrition_no['MonthlyIncome'], label='Attrition: No', marker='o')
plt.title('Average Monthly Income by Age and Attrition')
plt.xlabel('Age')
plt.ylabel('Average Monthly Income')
plt.legend()
plt.grid(True)
plt.show()
```
