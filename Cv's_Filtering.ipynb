```python
import os
import re
import fitz
def extract_text_from_pdf(pdf_path):
    text = ''
    try:
        doc = fitz.open(pdf_path)
        for page in doc:
            text += page.get_text()
    except Exception as e:
        print(f"Error reading PDF '{pdf_path}': {str(e)}")
    return text

pdf_dir = '/content/CVs'

keywords = ['AutoCAD', 'analysis', 'machine learning', 'JAVA', 'HTML', 'data', 'Unity', 'python', 'kotlin']

def normalize_text(text):
    return text.lower().replace(",", "")

for filename in os.listdir(pdf_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_dir, filename)
        cv_text = extract_text_from_pdf(pdf_path)
        preprocessed_text = normalize_text(cv_text)
        found_keywords = []
        for keyword in keywords:
            if normalize_text(keyword) in preprocessed_text:
                found_keywords.append(keyword)
        if found_keywords:
            print(f"Filename: {filename}, Keywords: {', '.join(found_keywords)}")
```

```python
experiences = profile_data['experiences']
```

```python
from datetime import datetime
total_years = 0
total_experiences = 0

for experience in experiences:
    start_date = datetime(experience['starts_at']['year'], experience['starts_at']['month'], experience['starts_at']['day'])
    if experience['ends_at'] is None:
        end_date = datetime.now()
    else:
        end_date = datetime(experience['ends_at']['year'], experience['ends_at']['month'], experience['ends_at']['day'])

    duration = end_date - start_date
    years_spent = duration.days / 365.25  # Considering leap years
    print("Years spent at", experience['company'], "{:.2f} years".format(years_spent))
    total_years += duration.days / 365.25  # Considering leap years
    total_experiences += 1

if total_experiences > 0:
    average_stay_years = total_years / total_experiences
    print("Average duration of staying at one place: {:.2f} years".format(average_stay_years))
else:
    print("No experiences found.")
```
